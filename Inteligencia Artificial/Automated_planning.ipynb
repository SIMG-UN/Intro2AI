# Automated Planning

El **planning** (planeacin) se define como la tarea de encontrar una secuencia de acciones para lograr un objetivo en un entorno discreto, determinista, esttico y completamente observable. Esta tarea es esencial para resolver problemas en diversas reas, desde la robtica hasta los videojuegos, donde el entorno es predecible y el objetivo debe alcanzarse siguiendo una secuencia de pasos planificados.

En **classical planning**, la idea bsica es encontrar una secuencia de acciones que lleven desde un estado inicial hasta un estado objetivo. Esto implica que todas las acciones deben ser especficas para el problema, deterministas, y el estado debe ser completamente observable. Este tipo de planeacin se basa en representar el entorno de forma factorizada y resolver el problema a travs de algoritmos de bsqueda.

Un ejemplo clsico es el ajedrez. Vamos a comparar dos escenarios para ilustrar cmo la complejidad de la planeacin vara segn el entorno y el objetivo.

### Problema 1: Llevar la reina a A8
* **Objetivo**: Llevar la reina desde su posicin inicial (por ejemplo, A1) a la casilla A8 en un tablero de ajedrez vaco.
* **Solucin**: La bsqueda en anchura o en profundidad son estrategias adecuadas en este caso, ya que el nmero de acciones posibles es reducido y el objetivo est claramente definido. **Explicacin**: Este problema se resuelve fcilmente con algoritmos de bsqueda en un espacio de estados pequeo y bien estructurado. La reina tiene un movimiento directo hacia A8, por lo que la planeacin se reduce a encontrar ese camino.

## Limitaciones de la Planeacin Clsica

Aunque los problemas de ajedrez ilustran bien el concepto de planeacin, tambin destacan algunas de las limitaciones de los mtodos clsicos de planeacin:

1. **Heursticas ad hoc**: En muchos problemas, se necesita conocimiento especfico del dominio para crear heursticas que guen la bsqueda. Cada nuevo dominio requiere que diseemos manualmente funciones de evaluacin, lo cual no es escalable ni flexible.
2. **Representacin de grandes espacios de estado**: Los mtodos clsicos tienden a representar el espacio de estados de manera explcita, lo que puede llevar a una explosin combinatoria. Por ejemplo, en el ajedrez, las combinaciones de posiciones y movimientos generan un espacio de bsqueda muy grande.

## Ejemplo de Representacin en Planeacin

En respuesta a las limitaciones de la planeacin clsica, los investigadores han desarrollado lenguajes de definicin de dominios de planeacin, como el **Planning Domain Definition Language (PDDL)**. Este lenguaje permite expresar las acciones, estados y objetivos de manera ms general y abstracta. En lugar de enumerar cada posible accin para cada estado, PDDL permite la definicin de esquemas de acciones que cubren mltiples casos de manera compacta.

Un estado en PDDL se representa como una conjuncin de "fluyentes" atmicos. Un **fluyente** es un aspecto del mundo que cambia con el tiempo, como la posicin de una pieza en el ajedrez o el nivel de energa de un vehculo. A travs de estos esquemas, es posible describir dominios y problemas de manera ms flexible y reutilizable.



# Planificacin Automatizada

La **planificacin automatizada** se define como la tarea de encontrar una secuencia de acciones para lograr un objetivo en un entorno discreto, determinista, esttico y completamente observable. Esta disciplina es fundamental en diversos campos, desde la robtica hasta los videojuegos, donde se requiere navegar en entornos predecibles para alcanzar metas especficas.

## Planificacin Clsica

En la **planificacin clsica**, el objetivo principal es determinar una serie de acciones que transformen un estado inicial en un estado objetivo. Este enfoque asume que:

1. Todas las acciones son especficas para el problema.
2. Las acciones son deterministas (producen siempre el mismo resultado).
3. El estado del entorno es completamente observable.

La planificacin clsica utiliza una representacin factorizada del entorno y emplea algoritmos de bsqueda para resolver problemas.

## Estrategias de Bsqueda en Planning

Existen dos enfoques principales para abordar problemas de planificacin:

### 1. Bsqueda Hacia Adelante

- **Proceso**: Comienza desde el estado inicial y explora todas las acciones posibles.
- **Ventajas**: 
  - Es directa y no requiere una descripcin detallada del objetivo final.
  - Flexible para problemas con objetivos poco definidos.
- **Desafos**: 
  - Puede ser ineficiente en problemas con muchos estados y acciones.
  - El espacio de bsqueda crece rpidamente.

### 2. Bsqueda Hacia Atrs

- **Proceso**: Inicia desde el estado objetivo y retrocede hacia el estado inicial.
- **Ventajas**: 
  - Puede ser ms eficiente si el objetivo est claramente definido.
  - Reduce el espacio de bsqueda en algunos casos.
- **Desafos**: 
  - Requiere que el objetivo sea especfico y bien definido.

## Complejidad y Desafos

La complejidad en la planificacin surge de:

1. El vasto nmero de estados posibles.
2. La cantidad de acciones disponibles en cada estado (factor de ramificacin).
3. La longitud de las secuencias de acciones necesarias para alcanzar el objetivo.

Estos factores pueden llevar a una explosin combinatoria, haciendo que incluso problemas aparentemente simples sean computacionalmente costosos.

## Heursticas y Optimizacin

Para manejar la complejidad, se emplean heursticas:

- **Definicin**: Estimaciones que indican qu tan cerca est un estado del objetivo.
- **Funcin**: Reducen el espacio de bsqueda al enfocar la exploracin en los estados ms prometedores.
- **Importancia**: Cruciales para resolver problemas complejos de manera eficiente.

## Representacin en Planificacin

Para abordar las limitaciones de la planificacin clsica, se han desarrollado lenguajes de definicin de dominios como el **Planning Domain Definition Language (PDDL)**:

- Permite expresar acciones, estados y objetivos de manera ms abstracta y general.
- Utiliza "fluyentes" para representar aspectos del mundo que cambian con el tiempo.
- Facilita la creacin de esquemas de acciones que cubren mltiples casos de manera compacta.
